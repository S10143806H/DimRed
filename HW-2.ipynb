```python
import matplotlib.pyplot as plt
import numpy as np

def g(x):
  return 50 + 30 * np.exp(-100 * x)

def f(x):
  return 120 * x + (1 - x) * g(x)

def normalize_f(f_x):
  # subtract the minimum value from f_x
  f_x = f_x - np.min(f_x)
  # divide f_x by the maximum value
  f_x = f_x / np.max(f_x)
  # return the normalized f_x
  return f_x

x = np.linspace(0, 0.33, 100)
g_x = g(x)
f_x = f(x)
f_star_x = normalize_f(f_x)

plt.plot(x, g_x, label="g(x)")
plt.plot(x, f_x, label="f(x)")
plt.xlabel("x")
plt.ylabel("g(x), f(x)")
plt.title("g(x)=50+30e^(-100x), f(x)=120x+(1-x)g(x)")
plt.legend()
plt.show()

# generate random 20 points for x
x = np.linspace(0, 1, 20)
x = np.around(x, decimals=3)
g_x = g(x)
f_x = f(x)

print("x")
print(x)
print("------------------------------------------------")
print("g_x")
print(g_x)
print("------------------------------------------------")
print("f_x")
print(f_x)
print("------------------------------------------------")

# normalize f(x)
f_star_x = normalize_f(f_x)
plt.plot(x, f_star_x, label="f*(x)")
plt.xlabel("x")
plt.ylabel("f*(x)")
plt.title("f*(x)")
plt.legend()
plt.show()
print(f_star_x)
```
